```python
from datasets import load_dataset
```

```python
dataset_test = load_dataset(
   'common_voice', 'ta', split="test")
```

```python
import os
import librosa
import soundfile
import pandas
import polyglot
from polyglot.text import Text
```

```python
files = list()
transcripts = list()

for file, transcript in zip(dataset_test["path"], dataset_test["sentence"]):
    file_name = file.split("/")[-1].replace("mp3", "wav")
    path = "/home/ubuntu/speech_data/t-seed/CommonVoice/Tamil"
    new_path = os.path.join(path, file_name)
    audio, sr = librosa.load(file, sr=16000)
    soundfile.write(new_path, audio, 16000)
    files.append(file_name)
    text = Text(transcript, hint_language_code="ta")
    transcripts.append(" ".join(list(text.transliterate("en"))))

dictionary = {"audios": files, "text": transcripts}
df = pandas.DataFrame.from_dict(dictionary)
df.to_csv("/home/ubuntu/speech_data/t-seed/CommonVoice/Tamil/test.tsv", sep="\t")
```
